#### Hypothesis
- H0 : There is no difference in the 'political_probability' between records with 'political' greater than 0 and 'political' equal to 0.
- H1 :  There is a difference in the 'political_probability' between records with 'political' greater than 0 and 'political' equal to 0. 

```python
from scipy import stats

group_political = df['political_probability'][df['political'] > 0]
group_not_political = df['political_probability'][df['political'] == 0]

t_statistic, p_value = stats.ttest_ind(group_political, group_not_political, equal_var=False)

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("The difference in political_probability is significant.")
else:
    print("There is not enough evidence to reject the null hypothesis.")
```
